```python
import numpy as np
import random
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

df = fetch_california_housing(as_frame=True)
x_1 = df.data["MedInc"].values
y = df.target.values.reshape(-1, 1)
n = len(x_1)
scaler = StandardScaler()
x_1 = scaler.fit_transform(x_1.reshape(-1, 1))

def OLS_optimiz(x, y):
  covm = np.cov(x.T, y.T)
  w = covm[0, 1] / covm[0, 0]
  b = np.mean(y) - (w * np.mean(x))
  return float(w), float(b)

def pred(w, b, x):
  return w * x + b

def GradientDescent(w_0, b_0, rate, m):
  global n, x_1
  for i in range(m):
    dw = 2/n * np.sum(x_1 * (pred(w_0, b_0, x_1) - y))
    db = 2/n * np.sum(pred(w_0, b_0, x_1) - y)
    w = w_0 - (rate * dw)
    b = b_0 - (rate * db)
    w_0 = w
    b_0 = b
  return w_0, b_0

plt.scatter(x_1, y)

w, b = OLS_optimiz(x_1, y)
plt.plot(x_1, pred(w, b, x_1))

print(OLS_optimiz(x_1, y))

w_1, b_1 = GradientDescent(0, 0, 0.1, 200)
plt.plot(x_1, pred(w_1, b_1, x_1))


def stochGD(w, b, rate, epochs, batch_size):
    global n, x_1, y
    for epoch in range(epochs):
        indices = np.random.permutation(n)
        for i in range(0, n, batch_size):
            batch_indices = indices[i : i + batch_size]
            x_batch = x_1[batch_indices]
            y_batch = y[batch_indices]
            dw = 0
            db = 0
            actual_batch_size = len(x_batch)
            for j in range(actual_batch_size):
                prediction = w * x_batch[j] + b
                error = prediction - y_batch[j]
                dw += x_batch[j] * error
                db += error
            dw = (2 / actual_batch_size) * dw
            db = (2 / actual_batch_size) * db
            w = w - (rate * dw)
            b = b - (rate * db)
    return w, b

w_2, b_2 = stochGD(0, 0, 0.1, 10, 150)
plt.plot(x_1, pred(w_2, b_2, x_1), c = "r")
```
